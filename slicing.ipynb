```python
slide = 10
def slice_mrc(img):
    for i in range(len(img.shape)):
        for j in range(img.shape[i]):
            if j<(img.shape[i]-9): 
                if i ==0:
                    im = img[j:j+10]
                    write_mrc('sliced'+str(i)+str(j)+'.mrc', im)
                if i ==1:
                    im = img[:,j:j+10,:]
                    write_mrc('sliced'+str(i)+str(j)+'.mrc', im)
                if i ==2:
                    im = img[:,:,j:j+10]
                    write_mrc('sliced'+str(i)+str(j)+'.mrc', im)
```
